## More functions

```python
df= pd.DataFrame({
    "name":['gautam','ansh','bhavika',np.nan],
    "toy":[np.nan,'batmobile','gun',np.nan],
    "born":[pd.NaT,pd.Timestamp("2003-12-14"),pd.NaT,pd.NaT],
})
df
```

```python
df.dropna()
```

```python
#it will not drop till all will be na
df.dropna(how='all')
```

```python
df.drop_duplicates(subset=['born'],keep='last')
```

```python
df.info()
```

```python
df['name'].value_counts(dropna=True)
```

```python
df.isnull()
```

```python
df.notnull()
```
